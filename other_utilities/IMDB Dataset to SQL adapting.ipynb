```python
# Found a reason to do some python :-)
# Link to dataset used https://www.kaggle.com/neha1703/movie-genre-from-its-poster
import pandas as pd
import os

cwd = os.getcwd()

data = pd.read_csv (cwd + '\imdb\MovieGenre.csv', encoding = "latin")
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
newDf = df[['imdbId','Title', 'IMDB Score', 'Genre', 'Poster']] 
samplesDf = newDf.sample(n=1) #500 Movies will do, for now
```

```python
import re, datetime, random

start_date = datetime.date(2021, 1, 1)
end_date = datetime.date(2021, 3, 1)

time_between_dates = end_date - start_date
days_between_dates = time_between_dates.days


for index, row in samplesDf.iterrows():
    findYear = re.search(r"\(([A-Za-z0-9_]+)\)", row['Title'])
    #If year is available
    if(findYear):    
        random_number_of_days = random.randrange(days_between_dates)
        random_date = start_date + datetime.timedelta(days=random_number_of_days)
        
        #Easier to follow
        title = row['Title'].split('(')[0].replace("'", "''") #Hyphen issues
        year = findYear.group(1)
        genre = row['Genre']
        rating = row['IMDB Score']
        imdbId = row['imdbId']
        imageLink = row['Poster']
        createdAt = random_date
        #Query building
        sql = "insert into movie (title, year, genre, rating, \"imdbId\", \"imageLink\", \"createdAt\") values {}"
        sql = sql.format("('%s', '%s', '%s', '%f', '%s', '%s', '%s');" % (title, year, genre, rating, imdbId, imageLink, createdAt))
        #Avoid corrupted data
        if not 'nan' in sql:
            print(sql)
       
    
    
```
